# Image Compression
### Using K-Means Algorithm

We will be using __skimage__ package for handling the image.

```python
from skimage import io,novice
import numpy as np
```

Importing our already prepared k-means algorithm. <br>
Since you have _.ipynb_ (notebook) extension, it is difficult to import.
[Click here]() to download the _.py_ code of the same.

```python
import kmeans
```

#### Reading and pre-processing the image.
The __io.imread__ will help us in importing the input image and convert it into a matrix representation.
The matrix is $3-dimensional$ of order height_of_image X width_of_image X 3.
That is each element contains Red/Green/Blue intensity value at each pixel.

```python
picture = novice.open('bird_small.png')
picture.show()
```

```python
image = io.imread('bird_small.png')
```

```python
rows = image.shape[0]
cols = image.shape[1]
image = image.reshape(rows*cols,3)
```

#### Calling the K-Means Algorithm.
( Theory about how image compression occurs in this method is explained in the session )

```python
centroids = kmeans.randomInit(image,16,len(image),3)
clusters = np.zeros(len(image),dtype=int)
(finalCentroids,finalClusters) = kmeans.kmeans_call(image,clusters,centroids,10,3)
```

#### Compression
Replace each pixel intensity value with its corresponding centroid

```python
for i in range(len(image)):
    image[i] = finalCentroids[finalClusters[i]]
```

##### Saving and Displaying compressed image

```python
converted = np.asarray(image) 
converted = converted.reshape(rows,cols,3)
io.imsave('converted_bird.png',converted)
comppicture = novice.open('converted_bird.png')
comppicture.show()
```

The original image was $32.2 KB$. The compressed image is about $9 KB$.
